```python
import math
import matplotlib.pyplot as plt

x = [0, 1, 2, 3]
y = [63, 39, 30, 26]

plt.bar(x, y, align='center')
plt.xticks(range(len(x)), ['One-GPU', '2 GPUs', '3 GPUs', '4 GPUs'])
plt.ylim((0, 70))
for i, v in enumerate(y):
    plt.text(x=i, y=v + 0.1, s=str(v) + ' s', ha='center')
plt.xlabel('Using number of GPU device')
plt.ylabel('Training time per epoch (second)')
plt.show()
plt.close()
```

```python
x = [8,16,32,64,128,256]
GPU_Efficiency = [0.91,0.93,0.96,0.97,0.98,1.0]
Accuracy = [0.954,0.945,0.893,0.742,0.502,0.277]

plt.plot(x,GPU_Efficiency, label="GPU_Efficiency")
plt.plot(x,Accuracy, label="Accuracy")
plt.xlabel('Batch Size')
plt.legend()
plt.title("one GPU")
plt.show()
plt.close()
```

```python

x = [8,16,32,64,128,256]
Average_GPU_Efficiency = [0.86,0.91,0.94,0.96,0.98,0.99]
Accuracy = [0.975,0.974,0.943,0.815,0.502,0.213]

plt.plot(x,Average_GPU_Efficiency, label="Average_GPU_Efficiency")
plt.plot(x,Accuracy, label="Accuracy")
plt.xlabel('Batch Size')
plt.legend()
plt.title("4 GPUs")
plt.show()
plt.close()
```
